```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Module 3, Assignment 17</h2>

<h3>Part 1: Read the mario data</h3>

```python
data_file = "mario_file.json"
with open(data_file, 'r') as mario_file:
    data = json.loads(mario_file.read())

print(f"Number of items: {len(data)}")

# Response
attempts = []

# Predictors
likes = []
boos = []
num_comments = []
for item in data:
    i_attempts, i_likes, i_boos, i_comments = item["attempts"], item["likes"], item["boos"], item["num_comments"]
    # Restrict each value to be less than upper_bound
    # Otherwise our plot is very visually sparse
    upper_bound = 10**5
    if i_attempts > upper_bound or i_likes > upper_bound or i_boos > upper_bound or i_comments > upper_bound:
        continue
    attempts.append(i_attempts)
    likes.append(i_likes)
    boos.append(i_boos)
    num_comments.append(i_comments)

attempts = np.array(attempts)
likes = np.array(likes)
boos = np.array(boos)
num_comments = np.array(num_comments)
```

<h3>Part 2: Plot Mario Data</h3>

```python
sorted_likes = likes.argsort()
likes = likes[sorted_likes]
like_sorted_attempts = attempts[sorted_likes]

font1 = {'family':'serif','color':'black','size':15}
font2 = {'family':'serif','color':'blue','size':20}
plt.xlabel("Likes", fontdict=font1)
plt.ylabel("Attempts", fontdict=font1)
plt.title("Likes vs Attempts", fontdict=font2)
plt.scatter(likes, like_sorted_attempts, s=.5)
plt.show()

sorted_boos = boos.argsort()
boos = boos[sorted_boos]
boo_sorted_attempts = attempts[sorted_boos]

font1 = {'family':'serif','color':'black','size':15}
font2 = {'family':'serif','color':'blue','size':20}
plt.xlabel("Boos", fontdict=font1)
plt.ylabel("Attempts", fontdict=font1)
plt.title("Boos vs Attempts", fontdict=font2)
plt.scatter(boos, boo_sorted_attempts, s=.5)
plt.show()

sorted_comments = num_comments.argsort()
num_comments = num_comments[sorted_boos]
comment_sorted_attempts = attempts[sorted_comments]

font1 = {'family':'serif','color':'black','size':15}
font2 = {'family':'serif','color':'blue','size':20}
plt.xlabel("Boos", fontdict=font1)
plt.ylabel("Attempts", fontdict=font1)
plt.title("Comments vs Attempts", fontdict=font2)
plt.scatter(num_comments, comment_sorted_attempts, s=.5)
plt.show()
```
