```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_file = "mario_file.json"
with open(data_file, 'r') as mario_file:
    data = json.loads(mario_file.read())

print(f"Number of items: {len(data)}")
```

```python
# Response
attempts = []

# Predictors
likes = []
boos = []
num_comments = []
for item in data:
    attempts.append(item["attempts"])
    likes.append(item["likes"])
    boos.append(item["boos"])
    num_comments.append(item["num_comments"])

attempts = np.array(attempts)
likes = np.array(likes)
boos = np.array(boos)
num_comments = np.array(num_comments)

plt.plot(likes, attempts)
# plt.plot(boos, attempts)
# plt.plot(num_comments, attempts)
plt.show()
```
